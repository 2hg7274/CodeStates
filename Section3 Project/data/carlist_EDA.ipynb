```python
# 결측값 확인
car_df.isna().sum()
```

```python
# 제조사 결측치들의 차종 확인
car_df[ car_df['제조사'].isna() ]['모델명'].unique()
```

```python
# Citroen과 Peugeot의 차량이 많은데 제조사가 적혀 있지 않은 것으로 파악
# 따라서 Citroen과 Peugeot의 제조사만 따로 결측치 대체 작업 
```

```python
make_index = car_df[ car_df['제조사'].isna() ].index
make_index
```

```python
for i in make_index:
    if "Peugeot" in car_df.iloc[i]['모델명']  :
        car_df.loc[i, '제조사'] = '푸조'
    elif "Citroen" in car_df.iloc[i]['모델명']:
        car_df.loc[i, '제조사'] = '시트로앵'
    else:
        pass
```

```python
car_df.isna().sum()
```

```python
# 제조사의 결측치 행 삭제
car_df = car_df.dropna(axis=0)
car_df.isna().sum()

```

```python
car_df = car_df.drop_duplicates(['모델명'], keep='first')
car_df.reset_index(drop=True)
```

```python
car_df.info()
```

```python
# 배기량, 공차중량, 등급, co2, 복합연비의 데이터 타입 바꾸기
## 그전에 CO2배출량의 값들 중 'g/km'인 값들이 있음 
## 즉 이런 값들은 결측값이므로 삭제를 진행하는 작업이 필요
```

```python
# 문자열에 숫자유무 확인 함수
def hasNumber(stringVal):
    return any(elem.isdigit() for elem in stringVal)
```

```python
# 숫자가 없는 값들의 인덱스를 가진 행 삭제 
print(car_df.shape)
for i in car_df[car_df['CO2배출량'].apply(hasNumber)==False].index:
    car_df = car_df.drop([i], axis=0)
print(car_df.shape)
car_df = car_df.reset_index(drop=True)
```

```python
def cc(data):
    data = data.str.replace(',','')
    data = data.str.replace('cc','')
    data = data.astype(int)
    return data

def kg(data):
    data = data.str.replace(',','')
    data = data.str.replace('kg','')
    data = data.astype(int)
    return data

def co2(data):
    data = data.str.replace('g/km', '')
    data = data.astype(float)
    return data

def km_l(data):
    if '㎞/ℓ' in data:
        data = data.replace('㎞/ℓ', '')
        data = float(data)
    elif 'km/kWh' in data:
        data = data.replace('km/kWh', '')
        data = float(data)
    return data
```

```python
car_df['배기량'] = cc(car_df['배기량'])

car_df['공차중량'] = kg(car_df['공차중량'])

car_df['CO2배출량'] = co2(car_df['CO2배출량'])

car_df['복합연비'] = car_df['복합연비'].apply(km_l)
```

```python
car_df.to_csv('/Users/hanhyeongu/Section3/Project/data/car_list_EDA.csv',index=False, encoding='utf-8-sig')
```
